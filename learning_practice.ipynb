```python
import matplotlib.pyplot as plt        # plotting the heat map of corr data.

def plot_df(df,size=11):
    corr=df.corr()
    fig,ax= plt.subplots(figsize=(size,size))
    ax.matshow(corr)
    plt.xticks(range(len(corr.columns)), corr.columns)
    plt.yticks(range(len(corr.columns)), corr.columns)
    
plot_df(df)
```

```python
# Simple Linear regression

# importing libraries
import pandas as pd

# importing the dataset
dataset=pd.read_csv("Data/Regression/Salary_Data.csv")
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:,1].values


# splitting the dataset into the train set and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.33, random_state = 42)

from sklearn.linear_model import LinearRegression
li_model = LinearRegression()
li_model.fit(X_train, y_train)
```

```python
# checking accuracy score of train dataset
print(li_model.score(X_train, y_train))
```

```python
y_predict = li_model.predict(X_test)
```

```python
y_predict
```

```python
# visualization of train dataset
import matplotlib.pyplot as plt

plt.scatter(X_train, y_train, color = 'blue')
plt.plot(X_train, li_model.predict(X_train), color = 'red')
plt.title("training set = salary vs Exp")
plt.xlabel("Years of Exp")
plt.ylabel("Salary")
plt.show()
```

```python
# visualization of test dataset
import matplotlib.pyplot as plt

plt.scatter(X_test, y_test, color = 'blue')
plt.plot(X_train, li_model.predict(X_train), color = 'red')
plt.title("training set = salary vs Exp")
plt.xlabel("Years of Exp")
plt.ylabel("Salary")

plt.show()
```

```python
# Multiple linear regression 
# importing libraries

import pandas as pd
import matplotlib.pyplot as plt

# importing dataset
dataset = pd.read_csv("Data/Regression/50_Startups.csv")

X = dataset.iloc[:, :-1].values
y = dataset.iloc[:,4].values

# Encoding the catagorical data
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer([("State" , OneHotEncoder(), [3])], remainder = 'passthrough')
X = ct.fit_transform(X)
```

```python
X
```

```python
# Avoiding the dummy variables trap

X = X[: , 1:]
```

```python
X
```

```python
# Splitting the data

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 25)
```

```python
# MlR
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#Predicting
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
# Visualization /3D plotting
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
x1 = X_test[:, [2]]
x2 = X_test[:, [3]]
ax.scatter(x1, x2, y_pred, c ='r', marker='o')
ax.set_xlabel("X1 Label")
ax.set_ylabel("X2 Label")
ax.set_zlabel("y_pred Label")
plt.show()
```

```python
# y_pred vs y_test
x_serial = list(range(1, len(y_pred)+1))
plt.scatter(x_serial, y_pred, color = 'red')
plt.scatter(x_serial, y_test, color = 'blue')
plt.title("y_pred vs y_test")
plt.xlabel("x_serial")
plt.ylabel("y_pred vs y_test")
plt.show()
```
